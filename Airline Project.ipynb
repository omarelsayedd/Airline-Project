Clean Data

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
#string felid 
dataset.drop(['carrier', 'carrier_name', 'airport', 'airport_name'], axis=1, inplace=True)
```

```python
dataset
```

```python
x = dataset.drop(['late_aircraft_delay'], axis=1, inplace=False)
y = dataset['late_aircraft_delay']
```

ML Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_test, x_train, y_test, y_train = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
train_score = model.score(x_train, y_train)
print(f'the acuuuracy of trainig:{round(train_score*100,2)}%')
```

```python
test_score = model.score(x_test, y_test)
print(f'the acuuuracy of testing:{round(test_score*100,2)}%')
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")
```
